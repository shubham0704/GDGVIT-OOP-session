## Same operators can have different meanings according to context

 Operator overloading can also be done in python by using these dunder methods<br>
 
 ```ex-
 for + we have __add__
 for - we have __sub__```

````python
# Suppose we want to add two point in a 2-D space
# define a Point type


class Point:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
    def __str__(self):
        return "({}, {})".format(self.x, self.y)
````

````python
result = Point(1,2) + Point(2,-3)
````

````python
# the default operation of add can be modified
# using operator overloading


class Point:
    
    
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
    
    
    def __str__(self):
        return "({}, {})".format(self.x, self.y)
    
    
    def __add__(self, otherPoint):
        new_x = self.x + otherPoint.x
        new_y = self.y + otherPoint.y
        return Point(new_x, new_y)
````

````python
result = Point(1,2) + Point(2,-3)
print ('Resultant Point - ', result)
````
